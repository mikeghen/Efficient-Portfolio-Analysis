```python
## Initialized Imports

import os
import requests
import pandas as pd
from pathlib import Path
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
import json

%matplotlib inline

load_dotenv()
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
# Format current date as ISO format
start = pd.Timestamp("2020-07-10", tz="America/New_York").isoformat()
end = pd.Timestamp("2021-07-10", tz="America/New_York").isoformat()
# Set timeframe to one day ('1D') for the Alpaca API
timeframe = "1D"
```

```python
# Set the tickers: Stocks
tickers = ["AXP", "AMGN", "AAPL", "BA", "CAT", "CSCO", "CVX", "GS", "HD", "HON", "IBM", "INTC", 
           "JNJ", "KO", "JPM", "MCD", "MMM", "MRK", "MSFT", "NKE", "PG", "TRV", "UNH", "CRM", 
           "VZ", "V", "WBA", "WMT", "DIS", "AGG", "SPY", "VTI", "VTV", "VOE", "VBR",
           "VEA", "VWO", "IEMG"]
# Get current closing prices tickers
df_draft_stock = alpaca.get_barset(
    tickers,
    timeframe,
    start = start,
    end = end
).df

# Pick the closing prices for the asset classes
df_portfolio_stock = pd.DataFrame()

# fetch closing prices
df_portfolio_stock["AXP"] = df_draft_stock["AXP"]["close"]
df_portfolio_stock["AMGN"] = df_draft_stock["AMGN"]["close"]
df_portfolio_stock["AAPL"] = df_draft_stock["AAPL"]["close"]
df_portfolio_stock["BA"] = df_draft_stock["BA"]["close"]
df_portfolio_stock["CAT"] = df_draft_stock["CAT"]["close"]
df_portfolio_stock["CSCO"] = df_draft_stock["CSCO"]["close"]
df_portfolio_stock["CVX"] = df_draft_stock["CVX"]["close"]
df_portfolio_stock["GS"] = df_draft_stock["GS"]["close"]
df_portfolio_stock["HD"] = df_draft_stock["HD"]["close"]
df_portfolio_stock["HON"] = df_draft_stock["HON"]["close"]
df_portfolio_stock["IBM"] = df_draft_stock["IBM"]["close"]
df_portfolio_stock["INTC"] = df_draft_stock["INTC"]["close"]
df_portfolio_stock["JNJ"] = df_draft_stock["JNJ"]["close"]
df_portfolio_stock["KO"] = df_draft_stock["KO"]["close"]
df_portfolio_stock["JPM"] = df_draft_stock["JPM"]["close"]
df_portfolio_stock["MCD"] = df_draft_stock["MCD"]["close"]
df_portfolio_stock["MMM"] = df_draft_stock["MMM"]["close"]
df_portfolio_stock["MRK"] = df_draft_stock["MRK"]["close"]
df_portfolio_stock["MSFT"] = df_draft_stock["MSFT"]["close"]
df_portfolio_stock["NKE"] = df_draft_stock["NKE"]["close"]
df_portfolio_stock["PG"] = df_draft_stock["PG"]["close"]
df_portfolio_stock["TRV"] = df_draft_stock["TRV"]["close"]
df_portfolio_stock["UNH"] = df_draft_stock["UNH"]["close"]
df_portfolio_stock["CRM"] = df_draft_stock["CRM"]["close"]
df_portfolio_stock["VZ"] = df_draft_stock["VZ"]["close"]
df_portfolio_stock["V"] = df_draft_stock["V"]["close"]
df_portfolio_stock["WBA"] = df_draft_stock["WBA"]["close"]
df_portfolio_stock["WMT"] = df_draft_stock["WMT"]["close"]
df_portfolio_stock["DIS"] = df_draft_stock["DIS"]["close"]
df_portfolio_stock["AGG"] = df_draft_stock["AGG"]["close"]
df_portfolio_stock["SPY"] = df_draft_stock["SPY"]["close"]
df_portfolio_stock["VTI"] = df_draft_stock["VTI"]["close"]
df_portfolio_stock["VTV"] = df_draft_stock["VTV"]["close"]
df_portfolio_stock["VOE"] = df_draft_stock["VOE"]["close"]
df_portfolio_stock["VBR"] = df_draft_stock["VBR"]["close"]
df_portfolio_stock["VEA"] = df_draft_stock["VEA"]["close"]
df_portfolio_stock["VWO"] = df_draft_stock["VWO"]["close"]
df_portfolio_stock["IEMG"] = df_draft_stock["IEMG"]["close"]


# Drop the time component of the date
df_portfolio_stock.index = df_portfolio_stock.index.date

df_portfolio_stock.reset_index(inplace=True)

df_portfolio_stock = df_portfolio_stock.rename(columns={
    "index": "Trade DATE"
})

# Save the data to CSV: "Output/Stocks.csv"
df_portfolio_stock.to_csv("Output/Stocks.csv", index = False)

df_portfolio_stock.set_index(pd.to_datetime(df_portfolio_stock["Trade DATE"],infer_datetime_format=True ), inplace=True)
```

```python
# Set the tickers: Bonds
tickers = ["VTIP", "AGG", "MUB", "BNDX", "EMB", "JPST", "GBIL", "SHV"]
# Get current closing prices 
df_draft_bonds = alpaca.get_barset(
    tickers,
    timeframe,
    start = start,
    end = end
).df

# Pick the closing prices for the asset classes
df_portfolio_bonds = pd.DataFrame()

# fetch closing prices
df_portfolio_bonds["VTIP"] = df_draft_bonds["VTIP"]["close"]
df_portfolio_bonds["AGG"] = df_draft_bonds["AGG"]["close"]
df_portfolio_bonds["MUB"] = df_draft_bonds["MUB"]["close"]
df_portfolio_bonds["BNDX"] = df_draft_bonds["BNDX"]["close"]
df_portfolio_bonds["EMB"] = df_draft_bonds["EMB"]["close"]
df_portfolio_bonds["JPST"] = df_draft_bonds["JPST"]["close"]
df_portfolio_bonds["GBIL"] = df_draft_bonds["GBIL"]["close"]
df_portfolio_bonds["SHV"] = df_draft_bonds["SHV"]["close"]


# Drop the time component of the date
df_portfolio_bonds.index = df_portfolio_bonds.index.date

df_portfolio_bonds.reset_index(inplace=True)

df_portfolio_bonds = df_portfolio_bonds.rename(columns={
    "index": "Trade DATE"
})

# Save the data to CSV: "Output/Bonds.csv"
df_portfolio_bonds.to_csv("Output/Bonds.csv", index = False)

df_portfolio_bonds.set_index(pd.to_datetime(df_portfolio_bonds["Trade DATE"],infer_datetime_format=True ), inplace=True)
```

```python
# Set the tickers: Crypto
tickers = ["BTC", "ADA", "ETH", "UNI"]
# Pick the closing prices for the asset classes
df_portfolio_crypto = pd.DataFrame()

for ticker in tickers:
    path= Path(f"Input/{ticker}_Historical_Data_one_year.csv")
    df=pd.read_csv(path, parse_dates=True, infer_datetime_format=True )
    start = pd.Timestamp("2020-07-10", tz="America/New_York").isoformat()
    if ticker=='BTC':
        df = df.rename(columns={"Date": "Trade DATE"})
        df_portfolio_crypto["Trade DATE"] = df["Trade DATE"]
        df_portfolio_crypto[ticker] = df["Price"]
    else:
        df_portfolio_crypto[ticker] = df["Price"]

#Save the data to CSV: "Output/Crypto.csv"
df_portfolio_crypto.to_csv("Output/Crypto.csv", index = False)

df_portfolio_crypto.set_index(pd.to_datetime(df_portfolio_crypto["Trade DATE"],infer_datetime_format=True ), inplace=True)
df_portfolio_crypto.drop(columns="Trade DATE", inplace=True)   
```

```python
# Set the tickers: Oil 
tickers = ["OIL", "OILK", "BNO", "COG"]
# Get current closing prices 
df_draft_oil = alpaca.get_barset(
    tickers,
    timeframe,
    start = start,
    end = end
).df

# Pick the closing prices for the asset classes
df_portfolio_oil = pd.DataFrame()

# fetch closing prices
df_portfolio_oil["OIL"] = df_draft_oil["OIL"]["close"]
df_portfolio_oil["OILK"] = df_draft_oil["OILK"]["close"]
df_portfolio_oil["BNO"] = df_draft_oil["BNO"]["close"]
df_portfolio_oil["COG"] = df_draft_oil["COG"]["close"]


# Drop the time component of the date
df_portfolio_oil.index = df_portfolio_oil.index.date

df_portfolio_oil.reset_index(inplace=True)

df_portfolio_oil = df_portfolio_oil.rename(columns={
    "index": "Trade DATE"
})

# Save the data to CSV: "Output/Oil.csv"
df_portfolio_oil.to_csv("Output/Oil.csv", index = False)

df_portfolio_oil.set_index(pd.to_datetime(df_portfolio_oil["Trade DATE"],infer_datetime_format=True ), inplace=True)
```

```python
# Set the tickers: Gold
tickers = ["GLD", "BAR", "UGL", "DBP", "ZI", "ZG"]
# Get current closing prices 
df_draft_gold = alpaca.get_barset(
    tickers,
    timeframe,
    start = start,
    end = end
).df


# Pick the closing prices for the asset classes
df_portfolio_gold = pd.DataFrame()

# fetch closing prices
df_portfolio_gold["GLD"] = df_draft_gold["GLD"]["close"]
df_portfolio_gold["BAR"] = df_draft_gold["BAR"]["close"]
df_portfolio_gold["UGL"] = df_draft_gold["UGL"]["close"]
df_portfolio_gold["DBP"] = df_draft_gold["DBP"]["close"]
df_portfolio_gold["ZI"] = df_draft_gold["ZI"]["close"]
df_portfolio_gold["ZG"] = df_draft_gold["ZG"]["close"]

# Drop the time component of the date
df_portfolio_gold.index = df_portfolio_gold.index.date

df_portfolio_gold.reset_index(inplace=True)

df_portfolio_gold = df_portfolio_gold.rename(columns={
    "index": "Trade DATE"
})

# Save the data to CSV: "Output/Gold.csv"
df_portfolio_gold.to_csv("Output/Gold.csv", index = False)

df_portfolio_gold.set_index(pd.to_datetime(df_portfolio_gold["Trade DATE"],infer_datetime_format=True ), inplace=True)
```

```python
# Clean and Concatenate data frames: Stock, Bond, Crypto
df_stock_bonds_crypto = pd.concat([df_portfolio_stock, df_portfolio_bonds, df_portfolio_crypto], axis= 'columns', join='inner')

## Identified and dropped null records in the combined portfolio
df_stock_bonds_crypto = df_stock_bonds_crypto.dropna()

## Dropped duplicates in the combined protfolio
df_stock_bonds_crypto = df_stock_bonds_crypto.drop_duplicates()
df_stock_bonds_crypto.drop(columns="Trade DATE", inplace=True)  

# Save the data to CSV: "Output/Stock_Bond_Crypto.csv"

df_stock_bonds_crypto.to_csv("Output/Stock_Bond_Crypto.csv")
```

```python
# Clean and Concatenate data frames: Stock, Crypto, Gold
df_stock_crypto_gold = pd.concat([df_portfolio_stock, df_portfolio_crypto, df_portfolio_gold], axis= 'columns', join='inner')

## Identified and dropped null records in the combined portfolio
df_stock_crypto_gold = df_stock_crypto_gold.dropna()

## Dropped duplicates in the combined protfolio
df_stock_crypto_gold = df_stock_crypto_gold.drop_duplicates()
df_stock_crypto_gold.drop(columns="Trade DATE", inplace=True)  

# Save the data to CSV: "Output/Stock_Crypto_Oil.csv"

df_stock_crypto_gold.to_csv("Output/Stock_Crypto_Gold.csv")
```
